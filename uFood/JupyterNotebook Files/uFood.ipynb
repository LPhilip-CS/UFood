```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# --- Initial Data Exploration ---

```python
## Importing .csv file
uFood = pd.read_csv(r'/Users/lijophilip/Desktop/Data Analytics/Projects/uFood/Resources/u_food_marketing.csv')
uFood.head()
```

```python
# Removing all duplicates rows
uFood.drop_duplicates(keep = False, inplace = True)
uFood.info()
```

# --- Data Cleaning ---

```python
uFood.head()
```

```python
# Combining kids and teens columns
uFood['Total_Children'] = uFood[['Kidhome', 'Teenhome']].sum(axis=1)
uFood.head()
```

```python
# Replace integer values in marital columns
uFood['marital_Divorced'] = uFood['marital_Divorced'].replace({1:5,0:0})
uFood['marital_Married'] = uFood['marital_Married'].replace({1:4,0:0})
uFood['marital_Single'] = uFood['marital_Single'].replace({1:3,0:0})
uFood['marital_Together'] = uFood['marital_Together'].replace({1:2,0:0})
uFood['marital_Widow'] = uFood['marital_Widow'].replace({1:1,0:0})
```

```python
# Combining all marital status into one column
uFood['Marital_Status'] = uFood[['marital_Divorced', 'marital_Married', 'marital_Single', 'marital_Together', 'marital_Widow']].sum(axis=1)
```

```python
uFood.head(20)
```

```python
uFood.info()
```

```python
# Replace integer values in education columns
uFood['education_2n Cycle'] = uFood['education_2n Cycle'].replace({1:1,0:0})
uFood['education_Basic'] = uFood['education_Basic'].replace({1:2,0:0})
uFood['education_Graduation'] = uFood['education_Graduation'].replace({1:3,0:0})
uFood['education_Master'] = uFood['education_Master'].replace({1:4,0:0})
uFood['education_PhD'] = uFood['education_PhD'].replace({1:5,0:0})
```

```python
# Combining all education status into one column
uFood['Education_Status'] = uFood[['education_2n Cycle', 'education_Basic', 'education_Graduation', 'education_Master', 'education_PhD']].sum(axis=1)
```

```python
uFood[uFood['Education_Status']!=0].head()
```

```python
uFood.info()
```

```python
# Combining all campaign columns into one
uFood['Accepted_Campaigns'] = uFood[['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp2', 'AcceptedCmp1', 'Response']].sum(axis=1)
```

```python
# Returning a boolean value if customer accepted a campaign
uFood['Accepted_Campaigns'] = (uFood['Accepted_Campaigns'] != 0).astype(int)
```

```python
uFood[uFood['Accepted_Campaigns']!=0].head()
```

```python
# Checking for any correlation
uFood.corr(method ='pearson')['Accepted_Campaigns'].sort_values(ascending = False)
```

```python
# Checking for age range
uFood['Age'].sort_values(ascending= True)
```

```python
# Creating age groups to better categorize
age_groups = [(23,30),(31,40),(41,50),(51,60),(61,70),(71,85)]

def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return("Unknown")

uFood['Age_Group'] = uFood['Age'].apply(assign_age_group)
```

```python
uFood[['Age', 'Age_Group']].head()
```

```python
age_order = ['23 - 30', '31 - 40', '41 - 50', '51 - 60', '61 - 70', '71 - 85']
sns.pointplot(data = uFood, x = 'Age_Group', y = 'Accepted_Campaigns', order = age_order)
plt.title('Accepted Campaigns Per Age Group')
```

```python
# Taking a percentage of accepted campaigns for each age group
counts = uFood['Age_Group'].value_counts()
```

# --- OVERALL FINDINGS ---
## INSIGHTS
### 1. Customer in the age range between 30-70 were spending more money. However, they are less likely to accept campaigns in higher volumes.
### 2. Catalog shoppers were more likely to accept campaigns, but in person shoppers were likely to spend more. (40% Catalog, 30% Store, 30% Web). Recommend a split between all 
### 3. Very high correlation between people with no kids (or less kids) and spending. (Focus on this demographic) 
### 4. Education plays no impact on campaigns. (Don't target this group)
### 5. Martial Status doesn't play a big part. Although married, single, and together customers tend to spend more money.

## PROFITABLE RECOMMENDATION
### - Target middle aged people who are high earners, with no kids. Target on different platforms (40% Catalog, 30% Store, 30% Web).

### - Focus on age range between 21-30, 70 and up. They statistically accepted campaigns at a higher rate.
